# Initial analysis
Здесь проведена простая первичная аналитика полученных результатов

```python
import os
from IPython.display import Image, display
base = os.path.join('..', 'data', 'processed')
paths = {
    'discount': os.path.join(base, 'discount_distribution.png'),
    'top_genres': os.path.join(base, 'top_genres.png')
}
for name, p in paths.items():
    if os.path.exists(p):
        print(f'Found {name} image: {p}')
        display(Image(filename=p))
    else:
        print(f'Image not found: {p}')
```

```python
import pandas as pd
from pathlib import Path
p = Path('..') / 'data' / 'processed' / 'steam_sales_clean.csv'
if not p.exists():
    print(f'Файл не найден')
else:
    df = pd.read_csv(p)
    print('Shape:', df.shape)
    display(df.head())

    if 'discount' in df.columns:
        discounts = pd.to_numeric(df['discount'], errors='coerce').dropna()
        if len(discounts) == 0:
            print('Нет значений')
        else:
            print('Discounts — count:', len(discounts))
            print('Mean discount:', discounts.mean())
            print('Median discount:', discounts.median())
            print('Discount value counts (top 10):')
            display(discounts.value_counts().head(10))
    else:
        print('Столбец discount отсутствует в данных.')
```
